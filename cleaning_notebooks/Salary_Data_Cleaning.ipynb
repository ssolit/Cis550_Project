```python
df = pd.read_csv("../data/Levels_Fyi_Salary_Data.csv")
df = df.drop(columns=["Race", "Education", 'Race_Asian', 'Race_White', 'Race_Two_Or_More', 'Race_Black', 'Race_Hispanic', 
                      'otherdetails', 'gender', 'Masters_Degree', 'Bachelors_Degree', 'Doctorate_Degree', 'Highschool',
                      'Some_College', 'rowNumber', 'dmaid'])
df
```

### Timestamp: convert to years

```python
df['year'] = df['timestamp'].apply(lambda x: pd.to_datetime(x).year)
df.drop(columns=["timestamp"], inplace=True)
df['year']
```

### Company name matches

```python
df["company"] = df["company"].apply(lambda x: str(x).upper())
fix_names = {"MSFT": "MICROSOFT"}
df["company"] = df["company"].apply(lambda x: fix_names[x] if x in fix_names.keys() else x)
df
```

### Fix Locations

```python
def parse_city(x): 
    x = x.split(",")
    if len(x) == 2: 
        city = x[0]
        return city
    return None

def parse_state(x): 
    x = x.split(",")
    if len(x) == 2: 
        state = x[1]
        if state[0] == " ": 
            return state[1:]
        else: 
            return state
    return None

df["city"] = df["location"].apply(parse_city)
df["state"] = df["location"].apply(parse_state)
df["country"] = "United States"
df
```
